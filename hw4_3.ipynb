```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 25
time = np.arange(1,n)
M = 10
phi = np.zeros((n-1,M))
temp = np.array([75,77,76,73,69,68,63,59,57,55,54,52,50,50,49,49,49,50,54,56,59,63,67,72])
for i in range(M):
    phi[:,i] = np.power(time, i)
plt.figure()
plt.plot(time, temp, color = 'k')

beta = np.linalg.pinv(phi) @ temp
n = int(24/0.01)
x = np.linspace(0, 24, n)
Phi = np.zeros((len(x), M))
for j in range(M):
    Phi[:,j] = np.power(x,j)
f = Phi @ beta

plt.plot(x,f)

plt.plot(time,temp,'o')
```

```python
from sklearn import linear_model
n = 24
M = 11
phi = np.zeros((n,M))
temp = np.array([75,77,76,73,69,68,63,59,57,55,54,52,50,50,49,49,49,50,54,56,59,63,67,72])
time = np.arange(1,n+1)
for i in range(M):
    phi[:,i] = np.power(time, i)
Phi = np.zeros((len(x), M))
for j in range(M):
    Phi[:,j] = np.power(x,j)

lam = 0.1
ENet = linear_model.ElasticNet(alpha=1.0, copy_X=True, l1_ratio=lam, max_iter=10**5,random_state=0, fit_intercept = True)
ENet.fit(phi, temp)  
beta_ENet = ENet.coef_
f_ENet = ENet.predict(Phi)
plt.plot(x,f_ENet)

Ridge = linear_model.Ridge(alpha=1.0).fit(phi,temp)
beta_Ridge = Ridge.coef_
f_Ridge = Ridge.predict(Phi)
plt.plot(x,f_Ridge)

beta_lse = np.linalg.pinv(phi) @ temp
f_lse = Phi @ beta_lse
plt.plot(x,f_lse)
plt.plot(time,temp,'.')

Lasso = linear_model.Lasso(alpha=1.0).fit(phi,temp)
f_Lasso = Lasso.predict(Phi)
plt.plot(x,f_Lasso)
```
